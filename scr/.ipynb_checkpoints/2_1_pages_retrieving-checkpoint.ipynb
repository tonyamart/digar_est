### Texts extraction

Example with small sample of texts dated 1852

```r
do_subset_search(searchterm = "", searchfile = "dorevol.txt", subset = subset_test, source = "pages")
```

```r
texts <- tibble(text = readLines("1852.txt"),
               PageID = str_extract(text, "^.*[[:space:]]<p>"))
```

```r
head(texts)
```

```r
texts <- texts %>% 
    filter(!is.na(PageID)) %>% 
    mutate(PageID = str_remove_all(PageID, "\\t|<p>")) %>% 
    mutate(text = str_remove(text, "^.*[[:space:]]<p>")) %>%
    select(PageID, text) %>%
    mutate(text = str_replace_all(text, "<p>|</p>", " "))

head(texts, 2)
```

```r
texts <- texts %>% 
    filter(!is.na(PageID)) %>% 
    mutate(PageID = str_remove_all(PageID, "\\t|<p>")) %>% 
    mutate(text = str_remove(text, "^.*[[:space:]]<p>")) %>%
    select(PageID, text) %>%
    mutate(text = str_replace_all(text, "<p>|</p>", " "))

head(texts, 2)
```
