#### Detecting languages

```r
test <- tibble(text = readLines("1852.txt"),
               PageID = str_extract(text, "^.*[[:space:]]<p>"))

test <- test %>% 
    filter(!is.na(PageID)) %>% 
    mutate(PageID = str_remove_all(PageID, "\\t|<p>")) %>% 
    mutate(text = str_remove(text, "^.*[[:space:]]<p>")) %>%
    select(PageID, text) %>%
    mutate(text = str_replace_all(text, "<p>|</p>", " "))

head(test, 2)
```

```r
test <- test %>% mutate(lang = sapply(text, textcat))
```
