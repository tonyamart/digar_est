```r
library(tidyverse)
```

```r
library(digar.txts, lib.loc = "~/R_pckg/")
```

```r
meta <- read.csv("test/dorevol_langs.csv")
str(meta)
```

```r
meta_pages <- read.delim("test/meta_rus_subset_pages.tsv", sep = "\t")
str(meta_pages)
```

```r
meta_full <- left_join(meta, meta_pages, by = "PageID")
str(meta_full)
```

```r
meta_rus <- meta_full %>% 
    mutate(lang_cln = str_replace_all(lang, "^(\\w+)-.*", "\\1")) %>% 
    filter(lang_cln == "russian")

unique(meta_rus$lang_cln)
str(meta_rus) #74k pages
```

```r
subset_rus <- read.delim(file = "test/all_issues_rus.tsv", sep = "\t")
str(subset_rus)
selected_meta <- subset_rus %>%
    select(DocumentID, keyid, attrs, DocumentTitle, pages, zippath_pages, basename_pages, n_pages)
```

```r
names(selected_meta)[names(selected_meta) == "DocumentID"] <- "docid"
str(selected_meta)
```

```r
meta_rus <- left_join(meta_rus, selected_meta, by = "docid")
str(meta_rus)
```

```r
do_subset_search(searchterm = "", 
                 searchfile = "dorevol_rus.txt", 
                 subset = meta_rus, 
                 source = "pages")
```

```r
texts <- tibble(text = readLines("dorevol_rus.txt"))

texts <- texts %>% mutate(PageID = str_extract(text, "^.*[[:space:]]<p>")) %>% 
    filter(!is.na(PageID)) %>% 
    mutate(PageID = str_remove_all(PageID, "\\t|<p>")) %>% 
    mutate(text = str_remove(text, "^.*[[:space:]]<p>")) %>%
    select(PageID, text) %>%
    mutate(text = str_replace_all(text, "<p>|</p>", " "))
```

```r
head(texts, 2)
```

```r
write.csv(texts, file = "test/texts_dorevol_all.csv")
```
